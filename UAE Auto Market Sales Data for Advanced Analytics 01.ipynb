# Explortary Data Analysis (EDA)

# Year

```python
#No. of car sells in a year trend
sns.set(rc={'figure.figsize':(25,15)})
a = sns.countplot(x= 'purchased_year', data = db)

for bars in a.containers:
    a.bar_label(bars)

# Set labels and title
a.set_xlabel('Year of Purchase')
a.set_ylabel('Count')
a.set_title('Number of Purchases per Year')
```

```python
#Last 20 Years total Sales 
a = db.groupby(['purchased_year'],as_index= False)['price'].sum().sort_values(by = 'price', ascending = False).head(20)
sns.set(rc={'figure.figsize':(20,10)})
b = sns.barplot(x = 'purchased_year', y ='price', data = a )

# Set labels and title
b.set_xlabel('Year of Purchase')
b.set_ylabel('Total Sales')
b.set_title('Total amount Sales per Year')
```

**From above graphs, we can say that 2023 was the best year in terms of sales and number of car sold.**

# Brand & Model

```python
#top 10 brands by units sold
top_brands = db['brand'].value_counts().head(10).index

# Filter the DataFrame to include only the top 10 brands
db_filtered = db[db['brand'].isin(top_brands)]

# Set the figure size
sns.set(rc={'figure.figsize':(15, 5)})

# Plot the countplot
a = sns.countplot(x='brand', data=db_filtered)

# Add labels to the bars
for bars in a.containers:
    a.bar_label(bars)

# Set labels and title
a.set_xlabel('Brand')
a.set_ylabel('No. of cars sold')
a.set_title('Number of cars sold by top 10 brands')
```

```python
#Top 10 highest selling car brands 
a = db.groupby(['brand'],as_index= False)['price'].sum().sort_values(by = 'price', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
a = sns.barplot(x = 'brand', y ='price', data = a )

# Set labels and title
a.set_xlabel('Brand')
a.set_ylabel('Total Sales')
a.set_title('Total Sales of cars based on Brand')
```

```python
#Top 10 highest selling car model 
b = db.groupby(['brand','model'],as_index= False)['price'].sum().sort_values(by = 'price', ascending = False).head(10)
sns.set(rc={'figure.figsize':(20,10)})
a = sns.barplot(x = 'brand', y ='price', data = b ,hue = 'model')

# Set labels and title
a.set_xlabel('Brand')
a.set_ylabel('Amount of Sales')
a.set_title('Total amount of sales based on Brand')
```

**From the above graph, we can say that Mercedes Benz sold highest units folowed by BMW. Mercedes Benz also have highest sales.**

# Engine Capacity & Horse Power
